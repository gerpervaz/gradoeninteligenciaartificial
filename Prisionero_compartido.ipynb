```python
#Instrucciones para crear un juego de dilema de priosionero con el que jugar.
#En este juego participan dos jugadores, uno coopera y otro calla.
#Si ambos cooperan, ambos reciben 5 puntos.
#Si ambos callan, ambos reciben 1 punto.
#Si un jugador coopera y otro calla, el que calla recibe 20 puntos y el que coopera 0 puntos.
#El jugador con la puntuación más baja gana.
#El juego se repite 5 veces.

#Import the random module.
import random

#Create a function to play the game.
def play_game():
    #Create a variable to hold the player's score.
    player_score = 0
    #Create a variable to hold the computer's score.
    computer_score = 0
    #Create a variable to hold the number of rounds played.
    rounds_played = 0
    #Create a while loop to play the game 5 times.
    while rounds_played < 5:
        #Create a variable to hold the player's choice.
        player_choice = input("Choose to cooperate or defect. ")
        #Create a variable to hold the computer's choice.
        computer_choice = random.choice(["cooperate", "defect"])
        #Create an if statement to determine the player's score.
        if player_choice == "cooperate" and computer_choice == "cooperate":
            player_score += 5
        elif player_choice == "defect" and computer_choice == "defect":
            player_score += 1
        elif player_choice == "cooperate" and computer_choice == "defect":
            player_score += 0
        elif player_choice == "defect" and computer_choice == "cooperate":
            player_score += 20
        #Create an if statement to determine the computer's score.
        if computer_choice == "cooperate" and player_choice == "cooperate":
            computer_score += 5
        elif computer_choice == "defect" and player_choice == "defect":
            computer_score += 1
        elif computer_choice == "cooperate" and player_choice == "defect":
            computer_score += 0
        elif computer_choice == "defect" and player_choice == "cooperate":
            computer_score += 20
        #Print the player's choice.
        print("You chose to " + player_choice + ".")
        #Print the computer's choice.
        print("The computer chose to " + computer_choice + ".")
        #Print the player's score.
        print("Your score is " + str(player_score) + ".")
        #Print the computer's score.
        print("The computer's score is " + str(computer_score) + ".")
        #Increment the number of rounds played.
        rounds_played += 1
    #Create an if statement to determine the winner.
    if player_score < computer_score:
        print("You win.")
    elif player_score > computer_score:
        print("You lose.")
```

```python
play_game()
```

```python
#Create a priosoner's dilemma game to play with.
#This is a game where two players can choose to cooperate or defect.
#If both players cooperate, they both get x points.
#If both players defect, they both get y point.
#If one player cooperates and the other defects, the defector gets z points and the cooperator gets 0 points.
#The game is played 5 times.
#Every time, x, y and z values are randomly generated between 0 and 20.
#The running player plays against the computer.
#The running player enters  C to cooperate or D to defect.
#The computer will randomly choose to cooperate or defect every time.
#The player with the lower cumulative score wins.
#The player with the higher cumulative score loses.
#The player with the same cumulative score ties.
#Create a variable player_score starting with 0  to hold the player's score.
#Create a variable computer_score starting with 0 to hold the computer's score.
#Create a variable rounds_played to hold the number of rounds played.
#Display the instructions at the beginning of the game.
#Display the round number.
#Create a line break to separate the rounds.
#Display the scores at the end of the game.
#Display the winner at the end of the game.

import random

#Create a function to play the game.
def play_game():
    #Create a variable player_score starting with 0  to hold the player's score.
    player_score = 0
    #Create a variable computer_score starting with 0 to hold the computer's score.
    computer_score = 0
    #Create a variable rounds_played to hold the number of rounds played.
    rounds_played = 0
    #Display the instructions at the beginning of the game.
    print("Welcome to the Prisoner's Dilemma Game!")
    print("You will play against the computer.")
    print("You will both choose to cooperate or defect.")
    print("If both players cooperate, they both get x points.")
    print("If both players defect, they both get y point.")
    print("If one player cooperates and the other defects, the defector gets z points and the cooperator gets 0 points.")
    print("The game is played 5 times.")
    print("Every time, x, y and z values are randomly generated between 0 and 20.")
    print("The running player plays against the computer.")
    print("The running player enters  C to cooperate or D to defect.")
    print("The computer will randomly choose to cooperate or defect every time.")
    print("The player with the lower cumulative score wins.")
    print("The player with the higher cumulative score loses.")
    print("The player with the same cumulative score ties.")
    print()
    print("Let's play!")
    print()
    #Create a while loop to play the game 5 times.
    while rounds_played < 5:
        #Create a variable x to hold the value of x.
        x = random.randint(0, 20)
        #Create a variable y to hold the value of y.
        y = random.randint(0, 20)
        #Create a variable z to hold the value of z.
        z = random.randint(0, 20)
        #Add 1 to rounds_played.
        rounds_played += 1
        #Display the round number.
        print("Round", rounds_played)
        print()
        #Display the values of x, y and z.
        print("x =", x)
        print("y =", y)
        print("z =", z)
        print()
        #Create a variable player_choice to hold the player's choice.
        player_choice = input("Do you want to cooperate or defect? Enter C to cooperate or D to defect: ")
        #Create a variable computer_choice to hold the computer's choice.
        computer_choice = random.choice(["C", "D"])
        #Create a variable player_points to hold the player's points.
        player_points = 0
        #Create a variable computer_points to hold the computer's points.
        computer_points = 0
        #Create an if statement to determine the player's points.
        if player_choice == "C" and computer_choice == "C":
            player_points = x
        elif player_choice == "D" and computer_choice == "D":
            player_points = y
        elif player_choice == "C" and computer_choice == "D":
            player_points = 0
        elif player_choice == "D" and computer_choice == "C":
            player_points = z
        #Create an if statement to determine the computer's points.
        if computer_choice == "C" and player_choice == "C":
            computer_points = x
        elif computer_choice == "D" and player_choice == "D":
            computer_points = y
        elif computer_choice == "C" and player_choice == "D":
            computer_points = 0
        elif computer_choice == "D" and player_choice == "C":
            computer_points = z
        #Add the player's points to the player's score.
        player_score += player_points
        #Add the computer's points to the computer's score.
        computer_score += computer_points
        #Display the player's choice.
        print("You chose", player_choice)
        #Display the computer's choice.
        print("The computer chose", computer_choice)
        #Display the player's points.
        print("You got", player_points, "points")
        #Display the computer's points.
        print("The computer got", computer_points, "points")
        #Display the player's score.
        print("Your score is", player_score)
        #Display the computer's score.
        print("The computer's score is", computer_score)
        #Create a line break to separate the rounds.
        print()
    #Display the scores at the end of the game.
    print("Your score is", player_score)
    print("The computer's score is", computer_score)
    #Create an if statement to determine the winner.
    if player_score < computer_score:
        print("You win!")
    elif player_score > computer_score:
        print("You lose!")
    elif player_score == computer_score:
        print("You tie!")
```

```python
#Call the play_game function.
play_game()
```

```python
# Se importa el paquete random necesario para que la máquinta tome resultados al azar.
import random
import logging

# Se fija el nivel de logging en DEBUG que es el más exigente y por lo tanto va a mostrar todos los mensajes.
# Además, se configura el formato incluyendo el momento en que ha ocurrido el momento del evento.
logging.basicConfig(level=logging.DEBUG, format=' %(asctime)s - %(levelname)s - %(message)s')
logging.info('Inicio del programa')

# Se crea la variable choices pra almacenar las elecciones del jugador.
choices = ["C", "D"]
logging.debug('Las posibles elecciones son: ' + str(choices))

# Se crea la función para jugar al juego.


def play_game():
    logging.info('Inicio del juego')

    # Se crea la variable player_score empezando en 0 para almacenar la puntuación del jugador.
    player_score = 0
    # Se crea la variable computer_score empezando en 0 para almacenar la puntuación de la máquina.
    computer_score = 0
    # Se crea la variable rounds_played para guardar el número de rondas jugadas.
    rounds_played = 0
    logging.debug("Se crea la variable player_score con valor de " + str(player_score))
    logging.debug("Se crea la variable computer_score con valor de " + str(computer_score))
    logging.debug("Se crea la variable rounds_played con valor de " + str(rounds_played))

    # Se muestran las instrucciones al comienzo del juego.
    print("¡Bienvenido al juego del Dilema del Prisionero! \nVas a jugar contra la máquina.")
    print("La policía tiene indicios para culparos a ti y a tu compinche de un delito.")
    print("Sin embargo les faltan pruebas para una condena rotunda. Por eso os han llevado a salas de interrogatorio "
          "distintas,")
    print("para sacaros una confesión.")
    print("Ahora debes decidir si cooperar o callar. Tu abogado te explica las consecuencias de una u otra acción...")
    print("Con lo que la policía sabe, aunque los dos calleis os van a condenar a 'y' años de carcel.")
    print("Si cooperas y tu compiche también, a ambos os caerán 'x' años.")
    print(
        "Pero si uno coopera y el otro calla, el que coopera sale libre (0 años), y al que calla le condenan a 'z' "
        "años.")
    print("Cuando quieras cooperar elige C (cooperate) y cuando quieras callar elige D (defect).")
    print(
        "El juego se juega al mejor de 5 rondas. Las condiciones penales cambian en cada ronda. Gana quien menos años "
        "de condena acumule")
    print("¡A jugar!")
    print()
    # Usamos un while loop para jugarlo 5 veces.
    # Alternativamente se podría haber usado 'for i in range(5):' o 'for i in range(1,1,6)'.
    while rounds_played < 5:
        # Añadimos 1 a la variable que contará el número de rondas.
        rounds_played += 1
        logging.debug("Inicio de la ronda: " + str(rounds_played))
        # Cada vez, los valores de x, y y z se generan al azar entre los naturales 1 y 20.
        # Creamos la variable x para almacenar los valores de x.
        x = random.randint(1, 20)
        # Creamos la variable x para almacenar los valores de y.
        y = random.randint(1, 20)
        # Creamos la variable x para almacenar los valores de z.
        z = random.randint(1, 20)
        # Punto de log.
        logging.debug("Primer valor x: " + str(x))
        logging.debug("Primer valor y: " + str(y))
        logging.debug("Primer valor z: " + str(z))
        # Para que tenga coherencia con el enunciado, debemos cumplir que z>x>y.
        # Alternativamente 'while z <= x or x <= y:' o 'while z <= x <= y:'.
        while not(z > x > y):
            x = random.randint(1, 20)
            y = random.randint(1, 20)
            z = random.randint(1, 20)
            # Punto de log.
            logging.debug("Nuevo valor de x: " + str(x))
            logging.debug("Nuevo valor de y: " + str(y))
            logging.debug("Nuevo valor de z: " + str(z))

        # Antes de cada ronda se muestra el número de ronda que llevamos.
        print("Ronda", rounds_played)
        print("----------")
        print()
        # Mostramos los valores de x, y y z con los que tienen que tomar una decisión.
        print("x =", x)
        print("y =", y)
        print("z =", z)
        print()

        # Usamos un bloque de try/except para evitar que el jugador escriba un valor que no sea C/c o D/d en la
        # entrada. Si el jugador entra un valor erroneo, se imprimirá un mensaje de error y se le pedirá que vuelva a
        # introducir un valor correcto.
        while True:
            try:
                # Se crea la variable player_choice para guardar la elección del jugador. El upper mejor lo ponemos
                # en el input porque si no hay que ponerlo en todos los if en forma de player_choice.upper
                #  y aquí solo se necesita poner una vez.
                player_choice = input(
                    "¿Deseas cooperar o callar? Introduce la letra C para cooperar, o D para callar: ").upper()
                logging.debug("El jugador elige " + player_choice)
                if player_choice not in choices:
                    raise ValueError
                # El break interrumpe la iteración en el momento en que deje de producirse el error.
                break
            except ValueError:
                logging.exception("ValueError exception")
                print("Has introducido un valor incorrecto. Por favor, introduce C/c para cooperar o D/d para callar.")
                # Si se añadiese aquí la palabra 'raise' otra vez se mostraría el nombre del error.

        # Se crea la variable computer_choice para guardar la elección de la máquina.
        computer_choice = random.choice(["C", "D"])
        logging.debug("El jugador eligió la entrada válida " + player_choice)
        logging.debug("La máquina eligió " + computer_choice)

        # Se crea (ronda 1) o resetea (>1 ronda) la variable player_points para guardar los años de cárcel del jugador.
        player_points = 0
        logging.debug("Valor de creación de la variable player_points: " + str(player_points))

        # Se crea (si ronda 1) o resetea (si ronda >1) la variable player_points para guardar los años de cárcel de
        # la máquina.
        computer_points = 0
        logging.debug("Valor de creación de la variable computer_points: " + str(computer_points))

        # Se crea una condición if para determinar las condenas del jugador y la máquina.
        if player_choice == "C" and computer_choice == "C":
            player_points = x
            computer_points = x
        elif player_choice == "D" and computer_choice == "D":
            player_points = y
            computer_points = y
        elif player_choice == "C" and computer_choice == "D":
            player_points = 0
            computer_points = z
        elif player_choice == "D" and computer_choice == "C":
            player_points = z
            computer_points = 0
        logging.debug("Resultado de la variable player_points: " + str(player_points))
        logging.debug("Resultado de la variable computer_points: " + str(computer_points))

        # Se añaden los años de cárcel al acumulado del jugador.
        player_score += player_points
        # Se añaden los años de cárcel al acumulado de la máquina.
        computer_score += computer_points
        logging.debug("Resultado de la variable player_score: " + str(player_score))
        logging.debug("Resultado de la variable computer_score: " + str(computer_score))
        # Se muestra la elección del jugador.
        print(f"Tú elegiste {player_choice}")
        # Se muestra la elección del ordenador.
        print(f"La máquina eligió {computer_choice}")
        # Se muestra la condena del jugador en la presente ronda.
        print(f"Eres condenado a {player_points} años de cárcel")
        # Se muestra la condena de la máquina en la presente ronda.
        print(f"La máquina es condenada a {computer_points} años de cárcel")
        # Se muestra el acumulado del jugador.
        # Alternativa: print("Acumulas", player_score, "años de cárcel")
        print(f"Acumulas {player_score} años de cárcel")
        # Se muestra el acumulado de la máquina.
        # Alternativa: print("La máquina acumula", computer_score, "años de cárcel")
        print(f"La máquina acumula {computer_score} años de cárcel")
        # Salto de línea para separar las rondas.
        print()
        logging.debug("Fin de la ronda número :" + str(rounds_played))

    # Se muestran los resultados al finalizar el juego.
    logging.info("Fin del juego")
    print("Has acumulado", player_score, "años de condena")
    print("La máquina ha acumulado", computer_score, "años de condena")
    # Se crea una condición if para determinar el ganador.
    if player_score < computer_score:
        print("¡Bien hecho!¡Has ganado!")
    elif player_score > computer_score:
        print("Lo sentimos, has perdido")
    elif player_score == computer_score:
        print("¡Has empatado con la máquina!")

# Call the play_game function.


play_game()
```
