```python
"""
Programa elaborado para la clase de programación I por Daniel Cortizo, Germán Pérez y Manuel Varela
en el que se propone un juego del dilema del prisionero frente a la máquina.
En la primera celda se encuentra la sintaxis lista para su juego, y en la segunda la misma sintaxis incluyendo
logs para realizar debugging.
"""

# Se importa el paquete random necesario para que la máquina pueda tomar valores al azar.
import random

# Se crea la lista 'choices' para almacenar las elecciones del jugador.
choices = ["C", "D"]

# Se crea la función para jugar al juego.


def play_game():
    # Se crea la variable player_score empezando en 0 para almacenar la puntuación del jugador.
    player_score = 0
    # Se crea la variable computer_score empezando en 0 para almacenar la puntuación de la máquina.
    computer_score = 0
    # Se crea la variable rounds_played para guardar el número de rondas jugadas.
    rounds_played = 0

    # Se muestran las instrucciones al comienzo del juego.
    print("¡Bienvenido al juego del Dilema del Prisionero! \nVas a jugar contra la máquina.")
    print("La policía tiene indicios para culparos a ti y a tu compinche de un delito.")
    print("Sin embargo, les faltan pruebas para una condena rotunda. Por eso os han llevado a salas de interrogatorio "
          "distintas, \npara sacaros una confesión.")
    print("Ahora debes decidir si cooperar o callar. Tu abogado te explica las consecuencias de una u otra acción...")
    print("Con lo que la policía sabe, aunque los dos calléis os van a condenar a 'y' años de cárcel.")
    print("Si cooperas y tu compiche también, a ambos os caerán 'x' años.")
    print(
        "Pero si uno coopera y el otro calla, el que coopera sale libre (0 años), y al que calla le condenan a 'z' "
        "años de cárcel.")
    print("Cuando quieras cooperar elige C (cooperate) y cuando quieras callar elige D (defect).")
    print(
        "El juego se juega al mejor de 5 rondas. Las condiciones penales cambian en cada ronda. Gana quien menos años "
        "de condena acumule")
    print("¡A jugar!")
    print()
    
    # Usamos un while loop para jugarlo 5 veces.
    # Alternativamente se podría haber usado un 'for i in range(5):'.
    while rounds_played < 5:
        # Añadimos 1 a la variable que contará el número de rondas.
        rounds_played += 1
        # Cada vez, los valores de x, y y z se generan al azar entre los naturales 1 y 20.
        # Creamos la variable x para almacenar los valores de x.
        x = random.randint(1, 20)
        # Creamos la variable x para almacenar los valores de y.
        y = random.randint(1, 20)
        # Creamos la variable x para almacenar los valores de z.
        z = random.randint(1, 20)
        
        # Para que tenga coherencia con el enunciado, debemos cumplir que z>x>y.
        # Alternativamente se podría haber usado 'while z <= x or x <= y:' o 'while z <= x <= y:'.
        while not(z > x > y):
            x = random.randint(1, 20)
            y = random.randint(1, 20)
            z = random.randint(1, 20)
            
        # Antes de cada ronda se muestra el número de ronda que llevamos.
        print("Ronda", rounds_played)
        print("----------")
        print()
        # Mostramos los valores de x, y y z con los que tienen que tomar una decisión.
        print("x =", x)
        print("y =", y)
        print("z =", z)
        print()

        # Usamos un bloque de try/except para evitar que el jugador escriba un valor que no sea C/c o D/d en la
        # entrada. Si el jugador entra un valor erroneo, se imprimirá un mensaje de error y se le pedirá que vuelva a
        # introducir un valor correcto.
        while True:
            try:
                # Se crea la variable player_choice para guardar la elección del jugador. El upper mejor lo ponemos
                # en el input porque si no hay que ponerlo en todos los if en forma de player_choice.upper
                #  y aquí solo se necesita poner una vez.
                player_choice = input(
                    "¿Deseas cooperar o callar? Introduce la letra C para cooperar, o D para callar: ").upper()
                if player_choice not in choices:
                    raise ValueError
                # El break interrumpe la iteración en el momento en que deje de producirse el error.
                break
            except ValueError:
                print("Has introducido un valor incorrecto. Por favor, introduce C/c para cooperar o D/d para callar.")
                # Si se añadiese aquí la palabra 'raise' otra vez se mostraría el nombre del error.

        # Se crea la variable computer_choice para guardar la elección de la máquina.
        computer_choice = random.choice(["C", "D"])
        
        # Se crea (si ronda 1) o resetea (si ronda >1) la variable player_points para guardar los años de cárcel 
        # del jugador.
        player_points = 0

        # Se crea (si ronda 1) o resetea (si ronda >1) la variable player_points para guardar los años de cárcel de
        # la máquina.
        computer_points = 0

        # Se crea una condición if para determinar las condenas del jugador y la máquina.
        if player_choice == "C" and computer_choice == "C":
            player_points = x
            computer_points = x
        elif player_choice == "D" and computer_choice == "D":
            player_points = y
            computer_points = y
        elif player_choice == "C" and computer_choice == "D":
            player_points = 0
            computer_points = z
        elif player_choice == "D" and computer_choice == "C":
            player_points = z
            computer_points = 0

        # Se añaden los años de cárcel al acumulado del jugador.
        player_score += player_points
        # Se añaden los años de cárcel al acumulado de la máquina.
        computer_score += computer_points

        # Se muestra la elección del jugador.
        print(f"Tú elegiste {player_choice}")
        # Se muestra la elección del ordenador.
        print(f"La máquina eligió {computer_choice}")
        # Se muestra la condena del jugador en la presente ronda.
        print(f"Eres condenado a {player_points} años de cárcel")
        # Se muestra la condena de la máquina en la presente ronda.
        print(f"La máquina es condenada a {computer_points} años de cárcel")
        # Se muestra el acumulado del jugador.
        # Alternativa: print("Acumulas", player_score, "años de cárcel")
        print(f"Acumulas {player_score} años de cárcel")
        # Se muestra el acumulado de la máquina.
        # Alternativa: print("La máquina acumula", computer_score, "años de cárcel")
        print(f"La máquina acumula {computer_score} años de cárcel")
        # Salto de línea para separar las rondas.
        print()

    # Se muestran los resultados al finalizar el juego.
    print(f"Has acumulado {player_score} años de condena")
    print(f"La máquina ha acumulado {computer_score} años de condena")
    # Se crea una condición if para determinar el ganador.
    if player_score < computer_score:
        print("¡Bien hecho!¡Has ganado!")
    elif player_score > computer_score:
        print("Lo sentimos, has perdido")
    elif player_score == computer_score:
        print("¡Has empatado con la máquina!")


# Llamada a la función play_game para empezar a jugar.


play_game()
```
